```python
import random

# --- 1. Load the words using your provided method ---
# NOTE: Using 'name.txt' assuming it is accessible in the same location
# where this script is run. Adjust the path if necessary.
try:
    words = open('../data/name.txt', 'r').read().splitlines()
except FileNotFoundError:
    print("Error: Could not find 'name.txt'. Please check the file path.")
    words = []

if words:
    print(f"Total words loaded: {len(words)}")

# --- 2. Define special characters and storage ---
special_chars = ['%', '1', '2', '_', '@', '3', '0', '#', '$', '!', '&']
modified_names = []

# --- 3. Process each word/name ---
for original_name in words:
    name_length = len(original_name)

    if name_length > 0:
        # Randomly choose a special character
        char_to_insert = random.choice(special_chars)

        # Randomly choose an insertion index (0 to length, inclusive)
        insertion_index = random.randint(0, name_length)

        # Insert the character using string slicing
        modified_name = (
            original_name[:insertion_index] +
            char_to_insert +
            original_name[insertion_index:]
        )
        modified_names.append(modified_name)
    else:
        modified_names.append("")


# -----------------------------------------------------------------
# --- 4. SAVE THE DATA TO synthetic_data.txt (New Step) ---
# -----------------------------------------------------------------
if modified_names:
    # Use 'w' mode to write the file, overwriting it if it already exists
    with open("../data/synthetic_data.txt", "w") as outfile:
        # Use '\n'.join() to join all names in the list, ensuring each
        # is on a separate line before writing to the file
        outfile.write('\n'.join(modified_names))

    print("\n---------------------------------------------------------")
    print(
        f"Successfully saved {len(modified_names)} modified names to synthetic_data.txt")
    print("---------------------------------------------------------")
```
